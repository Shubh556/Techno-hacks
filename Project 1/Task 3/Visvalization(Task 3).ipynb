Removing Outliers from the data

```python
df.describe()
```

```python
IQR=3.30-2.80
IQR
```

```python
lower_limit = 2.80-1.5*IQR
upper_limit = 3.30 +1.5*IQR
```

```python
lower_limit,upper_limit
```

```python
df[(df.SepalWidthCm<lower_limit)|(df.SepalWidthCm>upper_limit)]
```

```python
new_df_with_no_outliers=df[(df.SepalWidthCm>lower_limit)&(df.SepalWidthCm<upper_limit)]
new_df_with_no_outliers
```

```python
sns.boxplot(new_df_with_no_outliers[["SepalWidthCm","SepalLengthCm","PetalLengthCm","PetalWidthCm"]])
```

```python
plt.figure(figsize=(15,7))
sns.histplot(new_df_with_no_outliers["PetalLengthCm"], label="Petal Length",color="red")
sns.histplot(new_df_with_no_outliers["PetalWidthCm"], label="Petal Width",color="green")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(15,7))
sns.histplot(new_df_with_no_outliers["SepalLengthCm"], label="Sepal Length")
sns.histplot(new_df_with_no_outliers["SepalWidthCm"], label="Sepal Width")
plt.legend()
plt.show()
```
